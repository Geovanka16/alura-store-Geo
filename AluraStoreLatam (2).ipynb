## **5. Envío promedio por tienda**

```python
dataframes = {
    "Tienda 1": tienda,
    "Tienda 2": tienda2,
    "Tienda 3": tienda3,
    "Tienda 4": tienda4
}

resultados_costo_envio = []

for nombre, df in dataframes.items():

    costo_promedio = df['Costo de envío'].mean()

    resultados_costo_envio.append({
        'Tienda': nombre,
        'Costo Promedio de Envío (USD)': costo_promedio
    })

resultados_df = pd.DataFrame(resultados_costo_envio)

# Ordenamos de mayor a menor costo para identificar el más caro
resultados_ordenados = resultados_df.sort_values(by='Costo Promedio de Envío (USD)', ascending=False)


print(resultados_ordenados.to_string(formatters={'Costo Promedio de Envío (USD)': '${:.2f}'.format}, index=False))
```

##**6. Visualizaciones Clave del Proyecto Alura Store**

### **Gráfico de Barras**:
Comparación de Facturación Total
Este es el gráfico más importante para la decisión de venta, ya que muestra claramente que la Tienda 4 tiene la menor facturación.

```python
ingresos_data = {
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Ingreso Total': [1.150880e+09, 1.116344e+09, 1.098020e+09, 1.038376e+09]
}
df_ingresos = pd.DataFrame(ingresos_data).sort_values(by='Ingreso Total', ascending=False)

plt.figure(figsize=(9, 6))
bars = plt.bar(df_ingresos['Tienda'], df_ingresos['Ingreso Total'], color=['skyblue', 'skyblue', 'skyblue', 'red'])

plt.title('Facturación Total por Tienda (Análisis para Decisión de Venta)', fontsize=14)
plt.xlabel('Tienda')
plt.ylabel('Ingreso Total (en miles de millones)', fontsize=12)
plt.yticks(df_ingresos['Ingreso Total'], ['${:,.2f}B'.format(x / 1e9) for x in df_ingresos['Ingreso Total']])

# Resaltar la tienda con menor ingreso
bars[-1].set_color('salmon')
plt.text(bars[-1].get_x() + bars[-1].get_width()/2., bars[-1].get_height() + 5e6,
         "Menor Facturación", ha='center', color='red', weight='bold')

plt.tight_layout()
plt.show()
```

### **Gráfico de Dispersión (Scatter Plot): Evaluación vs. Facturación**
Este gráfico ayuda al Sr. Juan a ver si hay una relación entre cuánto se vende y qué tan satisfecho está el cliente, presentando los dos análisis clave en un solo lugar.

```python
datos_combinados = {
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Facturacion': [1.150880e+09, 1.116344e+09, 1.098020e+09, 1.038376e+09],
    'Evaluacion': [3.9767, 4.0373, 4.0483, 3.9958]
}
df_scatter = pd.DataFrame(datos_combinados)

plt.figure(figsize=(10, 6))

plt.scatter(df_scatter['Facturacion'], df_scatter['Evaluacion'],
            s=200, color='purple', alpha=0.7)

plt.title('Relación: Facturación vs. Satisfacción del Cliente', fontsize=14)
plt.xlabel('Facturación Total (en miles de millones)', fontsize=12)
plt.ylabel('Evaluación Promedio (Escala 1-5)', fontsize=12)

# Etiquetar los puntos para fácil identificación
for i in range(len(df_scatter)):
    plt.annotate(df_scatter['Tienda'][i],
                 (df_scatter['Facturacion'][i], df_scatter['Evaluacion'][i]),
                 textcoords="offset points", xytext=(5, -5), ha='left')

plt.xticks(df_scatter['Facturacion'], ['${:.2f}B'.format(x / 1e9) for x in df_scatter['Facturacion']], rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

### **Gráfico de Pastel (Pie Chart): Distribución de Categorías de la Tienda 4**
Este gráfico es ideal para mostrar la composición de las ventas de la Tienda 4 (nuestra candidata a la venta), ayudando a entender si se enfoca demasiado en una categoría específica o si tiene una distribución balanceada.

```python
dataframes = {
    "Tienda 1": tienda,
    "Tienda 2": tienda2,
    "Tienda 3": tienda3,
    "Tienda 4": tienda4
}

NOMBRE_COLUMNA_CATEGORIA = 'Categoría del Producto'


categorias_t4_conteo = tienda4[NOMBRE_COLUMNA_CATEGORIA].value_counts()

# Obtenemos el Top 5 de categorías
top_categorias_t4 = categorias_t4_conteo.head(5)

otros = categorias_t4_conteo.iloc[5:].sum()

serie_otros = pd.Series(data=[otros], index=['Otras Categorías'])

datos_pastel = pd.concat([top_categorias_t4, serie_otros])

# GENERACIÓN DEL GRÁFICO DE PASTEL
plt.figure(figsize=(8, 8))

plt.pie(datos_pastel.values,
        labels=datos_pastel.index,
        autopct='%1.1f%%',
        startangle=90,
        colors=plt.cm.Paired.colors)

plt.title('Distribución Porcentual de Ventas por Categoría (Tienda 4)', fontsize=14)
plt.tight_layout()
plt.show()
```

## **Informe final**



El presente informe tiene como objetivo evaluar el desempeño de las **cuatro tiendas de Alura Store** basándose en cinco métricas clave. La meta es determinar cuál de las tiendas debe ser vendida para permitir al Sr. Juan invertir en un nuevo negocio. El análisis concluyente recomienda vender la **Tienda 4**.

### 🔍 Insights Clave

* **Rendimiento Financiero (Facturación):** La **Tienda 4** registra el **menor ingreso total**, identificándola como la unidad de menor rendimiento.
* **Riesgo de Calidad (Evaluación):** La **Tienda 1** tiene la peor evaluación a pesar de ser la más rentable, lo que indica un riesgo potencial en la satisfacción del cliente.
* **Eficiencia Logística (Costo de Envío):** La **Tienda 4** es la más eficiente en costos de envío, lo que indica que su problema es la falta de **volumen** de ventas, no la operación.
* **Ventas por Producto/Categoría:** Los análisis de **Top 5 Productos** y **Distribución de Categorías** muestran que la Tienda 4 no logra generar el volumen masivo de ventas que sí alcanzan otras tiendas.

---

### Justificación Estratégica

La **Tienda 4** es la candidata más clara para la venta porque:

1.  **Es la unidad con el rendimiento financiero más débil.**
2.  Sus demás métricas (Evaluación y Costo de Envío) son estables, lo que significa que el problema es de mercado/volumen, y no un fallo operacional grave, permitiendo una desinversión limpia.

Se recomienda utilizar el capital de la venta para invertir en la **Tienda 3** (que tiene la mejor satisfacción de cliente) para aumentar su volumen de ventas y en la **Tienda 1** para mejorar la calidad y reducir el alto costo de envío.
